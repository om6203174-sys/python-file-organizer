```python
import os
import shutil

# Define the folder to organize (change this path to your Downloads folder)
FOLDER_TO_ORGANIZE = os.path.expanduser("~/Downloads")  # This gets your Downloads folder automatically

# Define file extensions for each category
IMAGE_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'}
VIDEO_EXTENSIONS = {'.mp4', '.mkv', '.mov', '.avi', '.wmv'}
DOCUMENT_EXTENSIONS = {'.doc', '.docx', '.txt', '.rtf'}
PDF_EXTENSIONS = {'.pdf'}

def create_folder_if_not_exists(folder_path):
    """Create a folder if it doesn't exist already"""
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"✅ Created folder: {folder_path}")
    else:
        print(f"📁 Folder already exists: {folder_path}")

def get_file_extension(filename):
    """Get the lowercase file extension from a filename"""
    return os.path.splitext(filename)[1].lower()

def organize_files():
    """Main function that organizes all files in the folder"""
    print(f"🔍 Scanning folder: {FOLDER_TO_ORGANIZE}")
    
    # Check if the main folder exists
    if not os.path.exists(FOLDER_TO_ORGANIZE):
        print(f"❌ Error: Folder '{FOLDER_TO_ORGANIZE}' does not exist!")
        return
    
    # Define the subfolder names and their extensions
    folders = {
        "Images": IMAGE_EXTENSIONS,
        "Videos": VIDEO_EXTENSIONS,
        "Documents": DOCUMENT_EXTENSIONS,
        "PDFs": PDF_EXTENSIONS
    }
    
    moved_files = 0  # Counter for moved files
    
    # Loop through all files in the folder
    for filename in os.listdir(FOLDER_TO_ORGANIZE):
        file_path = os.path.join(FOLDER_TO_ORGANIZE, filename)
        
        # Skip if it's a folder (we only want files)
        if os.path.isdir(file_path):
            print(f"⏭️  Skipping folder: {filename}")
            continue
        
        # Get the file extension
        extension = get_file_extension(filename)
        
        # Find which folder this file belongs to
        destination_folder = None
        for folder_name, extensions in folders.items():
            if extension in extensions:
                destination_folder = folder_name
                break
        
        # If file matches a category, move it
        if destination_folder:
            # Create the destination folder if needed
            dest_folder_path = os.path.join(FOLDER_TO_ORGANIZE, destination_folder)
            create_folder_if_not_exists(dest_folder_path)
            
            # Move the file
            dest_file_path = os.path.join(dest_folder_path, filename)
            
            # Handle duplicate filenames
            counter = 1
            original_dest = dest_file_path
            while os.path.exists(dest_file_path):
                name, ext = os.path.splitext(original_dest)
                dest_file_path = f"{name}_{counter}{ext}"
                counter += 1
            
            shutil.move(file_path, dest_file_path)
            print(f"📂 Moved '{filename}' → {destination_folder}/")
            moved_files += 1
        else:
            print(f"⏭️  Unknown file type: {filename}")
    
    print(f"\n🎉 Done! Moved {moved_files} files successfully!")

def main():
    """Main entry point - asks user for confirmation before organizing"""
    print("🚀 File Organizer")
    print("=" * 50)
    print(f"📂 Target folder: {FOLDER_TO_ORGANIZE}")
    print("\nThis will organize files into: Images, Videos, Documents, PDFs")
    
    # Ask for user confirmation
    confirm = input("\nDo you want to continue? (y/n): ").lower().strip()
    if confirm in ['y', 'yes']:
        organize_files()
    else:
        print("👋 Cancelled. No files were moved.")

if __name__ == "__main__":
    main()
```
